## Basic multi-armed bandit environment

```python
import numpy as np
class Environment():
    def __init__(self):
        self.machine_probs = np.array([0.4, 0.2, 0.1, 0.1, 0.1, 0.7])
        self.machine_rewards = np.array([1.0, 0.1, 2.0, 0.5, 6.0, 70.])
        self.n_actions = 6
    def step(self, a):
        reward = -25
        if np.random.uniform() < self.machine_probs[a]:
            reward += self.machine_rewards[a]
        return self.state, reward
    def reset(self):
        self.state = 0
        return self.state
    def render(self):
        print(self.state)
```
